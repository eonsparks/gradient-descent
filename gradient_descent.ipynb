```python
import numpy as np

loss = lambda w, y: ((w * 80 + 11.99) - y) ** 2
y = 81

ws = np.arange(-1, 3, .1)
losses = loss(ws, y)

plt.scatter(ws, losses)
plt.plot(1, loss(1, y), 'ro')
```

```python
gradient = lambda w, y: ((w * 80 + 11.99) - y) * 2
gradients = gradient(ws, y)
plt.scatter(ws, gradients)
plt.plot(1, gradient(1, y), "ro")
```
